```python
import pandas as pd
import numpy as nd
 
df = pd.read_csv('/content/train.csv')
```

```python
import matplotlib.pyplot as plt
sizes = df['Survived'].value_counts() 
fig, ax1 = plt.subplots() 
ax1.pie(sizes,labels = ['Not Survived',
'Survived'],autopct='#1.1f##',shadow=True)
plt.show()
```
